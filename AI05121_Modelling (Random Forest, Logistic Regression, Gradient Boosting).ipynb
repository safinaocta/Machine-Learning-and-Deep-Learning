# Preparation

Preparation biasa dilakukan untuk mempersiapkan data sebelum masuk dalam tahap pemodelan. <br>
Berikut adalah tahapan yang akan dilalui pada data `SC_HW1_bank_data.csv` sebelum tahap pemodelan :

## Import Library

```python
import pandas as pd
import numpy as np
```

## Input Dataset

```python
#Membaca data dan memasukkannya ke dalam bentuk Pandas Dataframe
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
```

```python
#Jalankan code untuk mengecek nama kolom yang tersedia
df.columns
```

```python
df.head()
```

```python
# Hilangkan kolom yang dirasa tidak relevan dengan model (contoh: ID). None dapat diisi dengan nama-nama kolom yang akan digunakan.
# Contoh df = df[['X1','X2', 'Y']].copy()

#START CODE
df = df[['Geography','Gender',	'Age',	'Tenure','Balance',	'NumOfProducts',	'HasCrCard',	'IsActiveMember',	'EstimatedSalary',	'Exited']].copy()
```

## Preprocessing

```python
df = pd.DataFrame(df)

# Memeriksa tipe data kolom-kolom
categorical_columns = df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
# Lakukan One-Hot Encoder untuk data categorical, dengan fungsi pandas get_dummies
df_encoded = pd.get_dummies(df, columns=categorical_columns)

# Menampilkan DataFrame yang telah diencode
df_encoded.head()
```

```python
#Pisahkan mana X (feature) dengan Y,
#Y adalah kolom "Exited"

#START CODE
X = df_encoded.drop(columns=['Exited'])  # Memisahkan semua kolom kecuali "Exited" sebagai fitur
y = df_encoded['Exited']  # Menggunakan kolom "Exited" sebagai target
```

```python
X.head()
```

```python
#Lakukan Scaler dan/atau Noermalisasi jika diperlukan
from sklearn.preprocessing import MinMaxScaler

#START CODE
scaler = MinMaxScaler()
X_transform = scaler.fit_transform(X)
```

```python
#Ini digunakan jika dilakukan scaler/Normalisas. Jika tidak, code ini bisa dilewat dan diganti dengan code yang ada di dalam komen
X_transform = pd.DataFrame(X_transform, columns = X.columns)
```

## Train-Test Split

```python
#Split menjadi train dan test dengan test_size 25%
#Tidak perlu mengubah code ini

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size = 0.25,random_state = 123)
```

# Modeling

## Model1


Model yang digunakan dalam contoh kode yang Anda berikan adalah **Logistic Regression**

dilakukan tuning parameter menggunakan GridSearchCV untuk mencari kombinasi hyperparameter terbaik dari model tersebut yang didefinisikan dalam param_grid, dan mencari yang terbaik berdasarkan metrik 'accuracy' dengan validasi silang (cross-validation) menggunakan 10 fold cross-validation. Hasil terbaik dari pencarian parameter akan diberikan oleh grid.best_params_.







```python
from sklearn.linear_model import LogisticRegression
```

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh
model1 = LogisticRegression()
params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}


#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(
             estimator= model1,
             param_grid= params,
             scoring = 'accuracy',
             n_jobs = 10, # core cpu yang digunakan
             cv = 10 # cross validation
            )

grid.fit(X_train,y_train)
grid.best_params_
```

```python
# Evaluasi Model
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

Model pertama dengan menggunakan Logistic Regression menghasilkan akurasi sebesar 81,36%

## Model2
### Soal :
Model yang digunakan adalah **Random Forest Classifier**  yang menggabungkan sejumlah besar pohon keputusan yang dikenal sebagai "pohon acak" (random trees).

jumlah estimator (n_estimators) : 100, 200, 300

kedalaman maksimum (max_depth) :  [None, 10, 20, 30]


min_samples_split : [2, 5, 10]


min_samples_leaf : [1,2,4]

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Inisialisasi model Random Forest Classifier
model2 = RandomForestClassifier()

# Definisikan daftar hyperparameter yang ingin Anda tuning
param_grid = {'n_estimators': [100, 200, 300],
              'max_depth': [None, 10, 20, 30],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [1, 2, 4]}

# Inisialisasi objek GridSearchCV
grid_search = GridSearchCV(estimator=model2, param_grid=param_grid,
                           scoring='accuracy', cv=10, n_jobs=-1)

# Melakukan pencarian hyperparameter terbaik
grid_search.fit(X_train, y_train)

# Menampilkan hyperparameter terbaik
best_params = grid_search.best_params_
print("Hyperparameter terbaik:", best_params)

# Model Random Forest Classifier dengan hyperparameter terbaik
best_model2 = grid_search.best_estimator_

# Melatih model Random Forest Classifier dengan hyperparameter terbaik
best_model2.fit(X_train, y_train)
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = best_model2.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

Akurasi yang diperoleh dengan model Random Forest Classifier adalah sebesar 86,12%

## Model3
Model ketiga yang digunakan adalah **Gradient Boosting Classifier**  

jumlah estimator (n_estimators) : [50, 100,200]

learning rate : [0.01, 0.1, 0.2]

max depth : [3, 4, 5]

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

# Inisialisasi model Gradient Boosting Classifier
model3 = GradientBoostingClassifier()

# Definisikan daftar hyperparameter yang ingin Anda tuning
params = {'n_estimators': [50, 100, 200],
          'learning_rate': [0.01, 0.1, 0.2],
          'max_depth': [3, 4, 5]}

# Inisialisasi objek GridSearchCV
grid_search = GridSearchCV(estimator=model3, param_grid=params,
                           scoring='accuracy', cv=10, n_jobs=-1)

# Melakukan pencarian hyperparameter terbaik
grid_search.fit(X_train, y_train)

# Menampilkan hyperparameter terbaik
best_params = grid_search.best_params_
print("Hyperparameter terbaik:", best_params)

# Melatih model Gradient Boosting Classifier dengan hyperparameter terbaik
best_model3 = grid_search.best_estimator_
best_model3.fit(X_train, y_train)
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = best_model3.predict(X_test)

print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print(accuracy_score(y_test, y_pred))
```

Menggunakan model Gradient Boosting Clasifier diperoleh akurasi sebeasar 86,92%

## Tarik Kesimpulan Model Mana yang terbaik beserta alasannya

Dari percobaan ketiga model di atas terlihat bahwa algoritma Gradient Boosting memiliki akurasi paling tinggi yaitu sebesar  86.92%., sehingga untuk memprediksi pelanggan bank yang nonaktif dapat menggunakan model algoritmaa Gradient Boosting
